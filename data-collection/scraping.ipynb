```python
import requests
from bs4 import BeautifulSoup

index = 'https://pandas.pydata.org/docs/reference/index.html'
response = requests.get(index)
soup = BeautifulSoup(response.content, 'html.parser')
links = [link.get('href') for link in soup.select('.toctree-l1 > a')]
links
```

```python
import os
os.makedirs('./data/pandas', exist_ok=True)
```

```python
internal_links = set()
for link in links:
  response = requests.get(f'https://pandas.pydata.org/docs/reference/{link}')
  soup = BeautifulSoup(response.content, 'html.parser')
  for link in soup.select('table a.internal'):
    href = link.get('href')
    internal_links.add(href)
internal_links = list(internal_links)
internal_links
```

```python
internal_links = list(set(link.split('#')[0] for link in internal_links if link != ''))
internal_links
```

```python
from tqdm import tqdm
```

```python
for link in tqdm(internal_links):
  if link.startswith('api/pandas.'):
    filename = link.split('/')[-1]
    url = f'https://pandas.pydata.org/docs/reference/{link}'
    response = requests.get(url)
    # download to html file
    with open(f'./data/pandas/{filename}', 'w') as f:
      f.write(response.text)
```

```python
pandas_data = []
for filename in tqdm(os.listdir('./data/pandas')):
  with open(f'./data/pandas/{filename}', 'r') as f:
    content = f.read()
    soup = BeautifulSoup(content, 'html.parser')
    item = {}
    q = soup.select_one('dd p')
    if q is not None:
      item['q'] = q.text
    a = soup.select_one('.rubric ~ div.doctest')
    if a is not None:
      item['a'] = a.text
    pandas_data.append(item)

pandas_data[:5]
```

```python
cleaned_data = []
for item in tqdm(pandas_data):
  if 'q' in item and 'a' in item:
    q = item['q'].strip()
    a = item['a'].strip()
    a = a.split('\n')
    lines = []
    for line in a:
      line = line.strip()
      if line.startswith('>>>') or line.startswith('...'):
        line = line[3:].strip()
        lines.append(line)
    if lines:
      cleaned_data.append({'q': q, 'a': '\n'.join(lines)})

cleaned_data[:5]
```

```python
import json
with open('./data/pandas.jsonl', 'w') as f:  
  for item in cleaned_data:
    f.write(json.dumps(item) + '\n')
```

```python
import requests
import time
import json

BASE_URL = "https://api.stackexchange.com/2.3"

TAGS = "data-science;python"
SITE = "stackoverflow"
PAGESIZE = 100
TARGET_COUNT = 10000

OUTPUT_FILE = "stackoverflow_data.json"

def fetch_questions(page):
    url = f"{BASE_URL}/search/advanced"
    params = {
        "page": page,
        "pagesize": PAGESIZE,
        "order": "desc",
        "sort": "creation",
        "tagged": TAGS,
        "answers": 1,
        "site": SITE,
        "filter": "withbody"
    }
    return requests.get(url, params=params).json()

def fetch_answers(question_ids):
    ids = ";".join(map(str, question_ids))
    url = f"{BASE_URL}/questions/{ids}/answers"
    params = {
        "order": "desc",
        "sort": "votes",
        "site": SITE,
        "filter": "withbody"
    }
    return requests.get(url, params=params).json()

def main():
    all_data = []
    page = 1

    while len(all_data) < TARGET_COUNT:
        print('\rFetching page', page, end='', flush=True)
        q_data = fetch_questions(page)

        if "items" not in q_data:
            break

        questions = q_data["items"]
        if not questions:
            break

        question_ids = [q["question_id"] for q in questions]
        a_data = fetch_answers(question_ids)

        answers_map = {}
        for ans in a_data.get("items", []):
            qid = ans["question_id"]
            answers_map.setdefault(qid, []).append(ans["body"])

        for q in questions:
            qid = q["question_id"]

            item = {
                "title": q["title"],
                "question": q["body"],
                "answers": answers_map.get(qid, [])
            }

            all_data.append(item)

            if len(all_data) >= TARGET_COUNT:
                break

        if "backoff" in q_data:
            time.sleep(q_data["backoff"])
        else:
            time.sleep(0.5)

        page += 1

    with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
        json.dump(all_data, f, ensure_ascii=False, indent=2)

if __name__ == "__main__":
    main()
```

```python
with open('./stackoverflow_data.json', 'r') as f:
    data = json.load(f)
data[:2]
```

```python
cleaned_data = []
for item in tqdm(data):
    q = item['title'] + '\n\n'
    soup = BeautifulSoup(item['question'], 'html.parser')
    for p in soup.select('p'):
        q += p.text + '\n'
    a = ''
    idx = 0
    while a == '' and idx < len(item['answers']):
        soup = BeautifulSoup(item['answers'][idx], 'html.parser')
        code = soup.select_one('pre > code')
        if code is not None:
            a = code.text.strip()
        idx += 1
    if q.strip() != '' and a.strip() != '':
        cleaned_data.append({'q': q.strip(), 'a': a.strip()})

cleaned_data[:5]
```

```python
import json
with open('./data/stackoverflow.jsonl', 'w') as f:
  for item in cleaned_data:
    f.write(json.dumps(item) + '\n')
```
