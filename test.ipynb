```python
import torch
from seq2seq import build_model, generate_code
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
ckpt = torch.load("models/seq2seq_prototype.pt", map_location=device)
stoi = ckpt["stoi"]
itos = ckpt["itos"]
model = build_model(len(itos), ckpt["emb_dim"], ckpt["hid_dim"], device)
model.load_state_dict(ckpt["model_state_dict"])
print(model.eval())
tests = [
    "how do i read a csv file in pandas",
    "sort the data by population column",
    "drop rows with missing values"
]
for q in tests:
    print("Q:", q)
    print("A:", generate_code(model, q, stoi, itos, device))
    print("-" * 40)
```
